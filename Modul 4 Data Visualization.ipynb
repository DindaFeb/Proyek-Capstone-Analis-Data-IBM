# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

# SQL query to get the age data from the database
QUERY = """
SELECT ConvertedComp 
FROM master
"""

# Execute the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Plot the histogram
plt.figure(figsize=(8,6))
plt.hist(df['ConvertedComp'], bins=10, color='lightgreen', edgecolor='darkblue')
plt.title('ConvertedComp Distribution')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# SQL query to get the raw age data from the database
QUERY = """
SELECT Age
FROM master
"""

# Execute the query and store the result in a dataframe
df = pd.read_sql_query(QUERY, conn)

# Check if there are any null or non-numeric values in the Age column
print(df['Age'].isnull().sum())  # Count of null values
print(df['Age'].dtype)  # Check the data type of the Age column
print(df['Age'].head())  # Inspect the first few rows of Age

# Convert to numeric and drop rows with invalid data
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # Convert non-numeric to NaN
df_clean = df.dropna(subset=['Age'])  # Drop NaN values

# Plot the boxplot using matplotlib with cleaned data
plt.figure(figsize=(6,8))
plt.boxplot(df_clean['Age'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add titles and labels
plt.title('Boxplot of Age')
plt.xlabel('Age')

# Show the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# SQL query to select Age and WorkWeekHrs from the database
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plotting Age vs. WorkWeekHrs
plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='blue')

# Add titles and labels
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Hours per Week')

# Show the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# SQL query to get WorkWeekHrs, CodeRevHrs, and Age data from the database
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""

# Execute the query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plotting WorkWeekHrs vs CodeRevHrs with Age as bubble size
plt.figure(figsize=(10,8))
bubble_size = df['Age'] * 20 

plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, color='blue')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble Size = Age)')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Load the entire table into a DataFrame for inspection


# SQL query to get the top 5 databases respondents wish to learn next year
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)

# Check the DataFrame columns to confirm
print(df.columns)

# Plotting the pie chart
plt.figure(figsize=(8,8))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Updated SQL query to get the median WorkWeekHrs and CodeRevHrs for age group 30 to 35
QUERY = """
WITH ranked_data AS (
    SELECT WorkWeekHrs, CodeRevHrs,
           ROW_NUMBER() OVER (PARTITION BY NULL ORDER BY WorkWeekHrs) AS rn,
           COUNT(*) OVER () AS total_count
    FROM master
    WHERE Age BETWEEN 30 AND 35
)
SELECT 
    AVG(WorkWeekHrs) AS median_WorkWeekHrs,
    AVG(CodeRevHrs) AS median_CodeRevHrs
FROM ranked_data
WHERE rn IN ((total_count + 1) / 2, (total_count + 2) / 2);
"""

df = pd.read_sql_query(QUERY, conn)
# Plotting the stacked chart
df.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age Group 30-35')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart

Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# SQL query to get the median ConvertedComp for each age from 45 to 60
QUERY = """
SELECT Age, 
       AVG(ConvertedComp) AS median_ConvertedComp
FROM (
    SELECT Age, ConvertedComp,
           ROW_NUMBER() OVER (PARTITION BY Age ORDER BY ConvertedComp) AS rn,
           COUNT(*) OVER (PARTITION BY Age) AS total_count
    FROM master
    WHERE Age BETWEEN 45 AND 60
) AS ranked_data
WHERE rn IN ((total_count + 1) / 2, (total_count + 2) / 2)
GROUP BY Age
ORDER BY Age;
"""

# Execute the query
df_median_comp = pd.read_sql_query(QUERY, conn)

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_median_comp['Age'], df_median_comp['median_ConvertedComp'], marker='o', linestyle='-', color='b')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.xticks(df_median_comp['Age'])  # Use the Age values as ticks
plt.grid()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# SQL query to get the count of each MainBranch
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
ORDER BY count DESC
"""

# Execute the query
df_mainbranch = pd.read_sql_query(QUERY, conn)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(df_mainbranch['MainBranch'], df_mainbranch['count'], color='pink')
plt.title('Count of Each Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```
